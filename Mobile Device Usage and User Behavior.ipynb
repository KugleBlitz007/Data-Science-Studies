Purpose: This dataset was designed primarily to support machine learning applications by providing insights into mobile user behavior. It enables predictive modeling of user engagement and consumption patterns based on device activity metrics.

Shape: The dataset contains 700 rows (user samples) and 11 columns (features).

Features:

User ID (Categorical): A unique identifier assigned to each user.

Device Model (Categorical): The model of the smartphone used.

Operating System (Categorical): The OS of the device (iOS or Android).

App Usage Time (Numerical): The total daily time spent using mobile applications, measured in minutes.

Screen On Time (Numerical): The average number of hours per day that the screen remains active.

Battery Drain (Numerical): The amount of battery consumed daily, measured in milliamp-hours (mAh).

Number of Apps Installed (Numerical): The total number of applications installed on the device.

Data Usage (Numerical): The daily mobile data consumption, measured in megabytes (MB).

Age (Numerical): The user's age.

Gender (Categorical): The gender of the user (Male or Female).

User Behavior Class (Categorical): A classification of user behavior based on usage patterns, with categories ranging from 1 (light usage) to 5 (extreme usage).

3) (r3) Our univariate analysis of the numerical age data—grouped into categories for 20s, 30s, 40s, and over 50 reveals that the largest populations fall within the 20s and 50s groups, while the 30s and 40s groups share roughly the same number of individuals. Below is the code that generated the grouped countplot of the visualization.

```python
'''
This is the youtube video that we used to take data from a csv file
https://www.youtube.com/watch?v=spALaS5BFX8

We also asked ChatGPT 4o to give us a way to access our csv file from a public github repo

We then asked ChatGPT 4o again to give us a way to group numerical data to it becomes categorical data

We used this website to generate a countplot : https://www.geeksforgeeks.org/countplot-using-seaborn-in-python/
'''

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Getting the csv file from our GitHub repo
csv_url = "https://raw.githubusercontent.com/KugleBlitz007/CSI4142/refs/heads/main/user_behavior_dataset.csv"
df = pd.read_csv(csv_url)

# We define the age grouping criteria
def categorize_age(age_group):
    if age_group < 29:
        return "20's"
    elif age_group < 39:
        return "30's"
    elif age_group < 49:
        return "40's"
    else:
        return "50+"

# We apply grouping to the dataset
df['Age Category'] = df['Age'].apply(categorize_age)

# Plotting the grouped-data countplot
plt.figure(figsize=(5,5))
sns.countplot(x="Age Category", data=df)
plt.xlabel("Age group of a user divided in 10 years")
plt.ylabel("Amount of users in that group")
plt.title("Distribution of user's age group by decade")
plt.grid(axis='y', linestyle='--')
plt.show()

```

4) (r4) Relatively speaking, iOS users exhibit more extreme usage patterns than Android users. We deduced this from a bivariate analysis of two categorical features, visualized using a pie chart. Here, “extreme usage” corresponds to Behavior Class groups 4 and 5, while “light usage” corresponds to groups 1 and 2.

```python
'''
This is the youtube video that we used to take data from a csv file
https://www.youtube.com/watch?v=spALaS5BFX8

We also asked ChatGPT 4o to give us a way to access our csv file from a public github repo

We asked ChatGPT 4o to give us a code that generates two pie chart at the same time

We used this website to generate a pie chart : https://www.w3schools.com/python/matplotlib_pie_charts.asp
'''

import pandas as pd
import matplotlib.pyplot as plt

# Getting the csv file from our GitHub repo
csv_url = "https://raw.githubusercontent.com/KugleBlitz007/CSI4142/refs/heads/main/user_behavior_dataset.csv"
df = pd.read_csv(csv_url)

# We Define grouping criteria for User Behavior Class
def categorize_behavior(behavior_class):
    if behavior_class in [1, 2]:
        return "Light"
    elif behavior_class in [3, 4]:
        return "Extreme"
    else:
        return None

# Apply grouping to the dataset
df['Behavior Category'] = df['User Behavior Class'].apply(categorize_behavior)

plt.figure(figsize=(10,10))

# Get unique operating systems
operating_systems = df["Operating System"].unique()

# We define a color map to categorize light users from extreme users
color_map = {
    "Light": "#1f77b4",
    "Extreme": "#d62728"
}

# This portion generates a pie chart for every mobile OS, here only Android and iOS
for i, os in enumerate(operating_systems, 1):
    plt.subplot(1, 2, i)
    os_data = df[df["Operating System"] == os]['Behavior Category'].value_counts()
    colors = [color_map[cat] for cat in os_data.index]
    os_data.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=colors)
    plt.title(f"User Behavior Category Distribution in {os}")
    plt.ylabel("")

plt.tight_layout()
plt.show()
```

5) (r5) Based on a bivariate analysis of the Device Model feature and user age groups (grouped by decades), we found that the Google Pixel 5 and Xiaomi Mi11 are the most popular smartphones among users in their 20s; Whereas, the Samsung Galaxy S21 and Google Pixel 5 are the least popular among users in their 30s and 40s (the working class). This is the code that generated a heatmap of this insight:

```python
'''
This is the youtube video that we used to take data from a csv file
https://www.youtube.com/watch?v=spALaS5BFX8

We reused all the code that ChatGPT 4o gave us so far

We used this website to generate a heatmap : https://www.geeksforgeeks.org/seaborn-heatmap-a-comprehensive-guide/
'''

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Getting the csv file from our GitHub repo
csv_url = "https://raw.githubusercontent.com/KugleBlitz007/CSI4142/refs/heads/main/user_behavior_dataset.csv"
df = pd.read_csv(csv_url)

# We define the age grouping criteria
def categorize_age(age_group):
    if age_group < 29:
        return "20's"
    elif age_group < 39:
        return "30's"
    elif age_group < 49:
        return "40's"
    else:
        return "50+"

# We apply grouping to the dataset
df['Age Category'] = df['Age'].apply(categorize_age)

# We create a contingency table for the heatmap
heatmap_data = pd.crosstab(df['Device Model'], df['Age Category'])

# We generate the heatmap
plt.figure(figsize=(5,5))
sns.heatmap(heatmap_data, annot=True, cmap="Blues", fmt="d")
plt.xlabel("Age Category")
plt.ylabel("Device Model")
plt.title("Heatmap of Device Model Across Age Categories")
plt.show()
```
